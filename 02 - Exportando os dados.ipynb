```python
import vaex
import pandas as pd
import geopandas as gpd
```

```python
df_iptu = vaex.open('data/IPTU-Distritos1995-2022-agrupados-por-distrito.hdf5').to_pandas_df()
gdf_distritos = gpd.read_file('data/SIRGAS_GPKG_distrito.gpkg')
```

```python
df_iptu.to_feather('data/IPTU-Distritos1995-2022-agrupados-por-distrito.feather')
```

```python
pd.read_feather('data/IPTU-Distritos1995-2022-agrupados-por-distrito.feather')
```

```python
df_iptu = df_iptu.merge(gdf_distritos.loc[:, ['ds_codigo', 'ds_nome']].astype({'ds_codigo': 'int'}), left_on='distrito', right_on='ds_codigo')
```

```python
df_pivot = df_iptu.pivot(index=['distrito', 'ds_nome'], values='Quantidade de Unidades', columns='ano')
```
